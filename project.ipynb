```python
data =  pd.read_csv("toxic_comments.csv")
```

```python
data.info() # информация о DF
```

```python
data = data.iloc[0:10]
```

```python
data.info()
```

```python
m = Mystem()
def lemmatize_clear(text):
   
    text = text.lower()
     
    lemm_text = "".join(m.lemmatize(text))
    text_clear = re.sub(r'[^а-яА-ЯёЁ]',' ', lemm_text)
            
    return " ".join(text_clear.split())
```

```python
data["text"] = data["text"].apply(lemmatize_clear)
```

```python
target = data['toxic']
features = data.drop(['toxic'], axis=1)
```

```python
features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.4, random_state=42)
```
